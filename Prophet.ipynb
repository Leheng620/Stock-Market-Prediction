```python
ticker_string = ["AAPL","ABB","ABBV","AEP","AGFS","AMGN","AMZN","BA","BABA","BAC","BBL",
         "BHP","BP","BRK-A","BUD","C","CAT","CELG","CHL","CHTR","CMCSA","CODI",
         "CSCO","CVX","D","DHR","DIS","DUK","EXC","FB","GD","GE","GOOG","HD",
         "HON","HRG","HSBC","IEP","INTC","JNJ","JPM","KO","LMT","MA","MCD","MDT","MMM",
         "MO","MRK","MSFT","NEE","NGG","NVS","ORCL","PCG","PCLN","PEP","PFE","PG",
         "PICO","PM","PPL","RDS-B","SNY","SO","SPLP","SRE","T",
         "TM","TOT","TSM","UL","UN","UNH","UPS","UTX","V","VZ","WFC","WMT","XOM"]

MSE_df = pd.DataFrame(columns=['Stock', 'MSE'])
Accuracy_df = pd.DataFrame(columns=['Stock','Accuracy'])

k=0
for ticker in ticker_string:
    data = pd.read_csv(f'/Users/liubohan/Downloads/Stock-Market-Prediction-main/price/{ticker}.csv')
    
    #Select columns Date and Close
    data = data[['Date','Adj Close']]
    data = data.rename({'Date':'ds','Adj Close':'y'},axis='columns')
    data_train = data.head(len(data)-60)

    #Prophet
    m = Prophet()
    m = Prophet(daily_seasonality=True)
    m.fit(data_train)
    future = m.make_future_dataframe(periods=60)#Periods
    forecast = m.predict(future)
    
    #Plot
    #figure1 = m.plot(forecast)
    #figure2 = m.plot_components(forecast)
    
    #Interative Plot
    #from prophet.plot import plot_plotly
    #plot_plotly(m,forecast)
    y_true = data['y'].tail(60)
    y_pred = forecast['yhat'].tail(60)
    
    #MSE
    MSE = mean_squared_error(y_true, y_pred)
    MSE_df.loc[k] = [ticker,MSE]
    
    #df = pd.DataFrame(y_pred.head(len(data['y'])))
    df = pd.DataFrame(forecast['yhat'])
    df.to_csv(f'/Users/liubohan/Downloads/Stock-Market-Prediction-main/prophet/{ticker}_prophet.csv')
    
    correct_pred = 0
    incorrect_pred = 0
    for i in range(1, len(y_true)):
        if (list(y_true)[i] - list(y_true)[i-1]) * (list(y_pred)[i] - list(y_true)[i-1]) > 0:
            correct_pred += 1
        else:
            incorrect_pred += 1
    Accuracy = correct_pred / (correct_pred + incorrect_pred)
    
    Accuracy_df.loc[k] = [ticker,Accuracy]
    k=k+1
    
    #if k==2:
    #    break
    
MSE_df.to_csv(r'/Users/liubohan/Downloads/Stock-Market-Prediction-main/prophet_mse_accuracy/mse.csv')
Accuracy_df.to_csv(r'/Users/liubohan/Downloads/Stock-Market-Prediction-main/prophet_mse_accuracy/accuracy.csv')
    
```
