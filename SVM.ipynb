```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
import random
import pandas as pd 
```

```python
tickerStrings = ["SPY", "AAPL", "TSLA", "AMZN", "META"]
def stocks_data(tickerStrings):
    for ticker in tickerStrings:
        df_temp = pd.read_csv(f'/content/drive/MyDrive/Colab Notebooks/ticker/ticker_{ticker}.csv', index_col='Date',parse_dates=True)
    return df_temp
df = stocks_data(tickerStrings).drop(axis = 1,labels ='ticker' ) # META
df = df[30:]
df.info()
df['Prediction'] = df[['Adj Close']].shift(-44)
print(df)
```

```python
#Create a data set X and convert it into numpy array , which will be having actual values
X = np.array(df.drop(['Prediction'],1))
#Remove the last 15 rows
X = X[:-44]
print(X)
```

```python
# Create a dataset y which will be having Predicted values and convert into numpy array
y = np.array(df['Prediction'])
# Remove Last 15 rows
y = y[:-44]
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html
# SVM Model
svr = SVR(kernel='linear', C=1.0, gamma=0.1)
# Train the model 
svr.fit(x_train, y_train)
```

```python
# The best possible score is 1.0
svm_confidence = svr.score(x_test, y_test)
print("svm confidence: ", svm_confidence)
```

```python
tickerStrings = ["SPY", "AAPL", "TSLA", "AMZN", "META"]
def stocks_data(tickerStrings):
    for ticker in tickerStrings:
        df_temp = pd.read_csv(f'/content/drive/MyDrive/Colab Notebooks/ticker/ticker_{ticker}.csv', index_col='Date',parse_dates=True)
    return df_temp
ticker = 'META'
df = stocks_data(tickerStrings).drop(axis = 1,labels ='ticker' ) # META
df = df[30:]
df.head()
df.info()
```

```python
ticker = 'META'
df = stocks_data(tickerStrings).drop(axis = 1,labels ='ticker' ) # META
class SVM:
    ticker = ''
    data = pd.DataFrame()
    def __init__(self, ticker,dataframe,traing_test_split_rate):
        self.ticker = ticker
        self.data = dataframe
        self.rate = traing_test_split_rate
        self.value = None
        self.__dataprocessing__()
    def __dataprocessing__(self,):
        # value表示涨跌, =1为涨，=0为跌
        self.value = pd.Series(self.data['Close'] - self.data['Close'].shift(1), index=self.data.index)
        self.value = self.value.bfill()
        self.value[self.value >= 0] = 1
        self.value[self.value < 0] = 0
        self.data['Value'] = self.value
        # 后向填充空缺值
        self.data = self.data.fillna(method='bfill')
        self.data = self.data.astype('float64')
        print(self.data.head())
        # 对样本特征进行归一化处理
        self.data = self.data.drop(['Value'], axis=1)
        self.data = preprocessing.scale(self.data)
    def predict(self,):
        L = len(self.data)
        train = int(L * self.rate)
        total_predict_data = L - train
        # 开始循环预测，每次向前预测一个值
        correct = 0
        train_original = train
        while train < L:
            Data_train = self.data[train - train_original:train]
            value_train = self.value[train - train_original:train]
            Data_predict = self.data[train:train + 1]
            value_real = self.value[train:train + 1]
            # 核函数分别选取'ploy','linear','rbf'
            # classifier = svm.SVC(C=1.0, kernel='poly')
            # classifier = svm.SVC(kernel='linear',C=1e3, gamma=0.1)
            classifier = svm.SVC(C=10.0, kernel='rbf')
            classifier.fit(Data_train, value_train)
            value_predict = classifier.predict(Data_predict)
            print("value_real=%d value_predict=%d" % (value_real[0], value_predict))
            # 计算测试集中的正确率
            if (value_real[0] == int(value_predict)):
                correct = correct + 1
            train = train + 1
        print(correct)
        print(total_predict_data)
        correct = correct * 100 / total_predict_data
        print("Correct=%.2f%%" % correct)

model = SVM(ticker,df,0.8)
model.predict()
```
